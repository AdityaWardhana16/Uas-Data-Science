```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import dataset
btc_data = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/BTC-USD-5Tahun2023.csv")

# Prepare Data
# Generate categories based on 'Date' (e.g., year bins)
btc_data['Year'] = pd.to_datetime(btc_data['Date']).dt.year
categories = btc_data['Year'].unique()
colors = [plt.cm.viridis(i / float(len(categories))) for i in range(len(categories))]

# Draw Plot for Each Year
plt.figure(figsize=(14, 8), dpi=100, facecolor='lightgray', edgecolor='k')

for i, year in enumerate(categories):
    subset = btc_data[btc_data['Year'] == year]
    plt.scatter(subset['Open'], subset['Close'], 
                s=15, c=[colors[i]], label=f"Year: {year}")

# Decorations
plt.gca().set(xlim=(0, btc_data['Open'].max() * 1.1), 
              ylim=(0, btc_data['Close'].max() * 1.1),
              xlabel='Open Price (USD)', ylabel='Close Price (USD)')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title("Scatterplot of BTC Open vs Close Prices (2018-2023)", fontsize=18, fontweight='bold')
plt.legend(fontsize=10, title="Years")
plt.grid(alpha=0.3)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import patches
from scipy.spatial import ConvexHull
import warnings

# Disable warnings
warnings.simplefilter('ignore')

# Set seaborn style
sns.set_style("white")

# Step 1: Import Data
btc_data = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/BTC-USD-5Tahun2023.csv")

# Prepare Data
btc_data['Year'] = pd.to_datetime(btc_data['Date']).dt.year
categories = np.unique(btc_data['Year'])

# Generate colors for each year
colors = [plt.cm.viridis(i / float(len(categories))) for i in range(len(categories))]

# Step 2: Draw Scatterplot for each Year
fig = plt.figure(figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')

for i, year in enumerate(categories):
    subset = btc_data[btc_data['Year'] == year]
    plt.scatter(subset['Open'], subset['Close'], 
                s=40, c=[colors[i]], label=f"Year {year}", edgecolors='black', linewidths=0.8)

# Step 3: Encircling for selected year (e.g., 2021)
def encircle(x, y, ax=None, **kw):
    if not ax:
        ax = plt.gca()
    p = np.c_[x, y]
    hull = ConvexHull(p)
    poly = plt.Polygon(p[hull.vertices, :], **kw)
    ax.add_patch(poly)

# Select data for encircling, e.g., Year 2021
btc_encircle_data = btc_data[btc_data['Year'] == 2021]

# Draw polygon around selected data
encircle(btc_encircle_data['Open'], btc_encircle_data['Close'], ec="crimson", fc="none", linewidth=2)
encircle(btc_encircle_data['Open'], btc_encircle_data['Close'], ec="gold", fc="none", linewidth=1.5)

# Step 4: Decorations
plt.gca().set(xlim=(0, btc_data['Open'].max() * 1.1), ylim=(0, btc_data['Close'].max() * 1.1),
              xlabel='Open Price (USD)', ylabel='Close Price (USD)')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("BTC Open vs Close Prices with Encircling (2021)", fontsize=22)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import new dataset
btc_data = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/BTC-USD-5Tahun2023.csv")

# Extract Year and Month from the 'Date' column
btc_data['Month'] = pd.to_datetime(btc_data['Date']).dt.month
btc_data['Year'] = pd.to_datetime(btc_data['Date']).dt.year

# Calculate the monthly average Close price for each Year
btc_monthly_avg = btc_data.groupby(['Year', 'Month'])['Close'].mean().reset_index()

# Rank the data by the average Close price for each Year and Month
btc_monthly_avg['Rank'] = btc_monthly_avg.groupby('Year')['Close'].rank(ascending=False)

# Sort by Rank
btc_monthly_avg_sorted = btc_monthly_avg.sort_values(by=['Year', 'Rank'])

# Draw Ranking Plot (Rank vs Month for each Year)
plt.figure(figsize=(16, 10), dpi=80)
sns.scatterplot(x='Month', y='Rank', hue='Year', data=btc_monthly_avg_sorted, palette='viridis', s=100, edgecolor='black')

# Decorations
plt.title('Bitcoin Monthly Ranking by Average Close Price', fontsize=22)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Rank (1 = Highest)', fontsize=14)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year', fontsize=12)
plt.gca().invert_yaxis()  # Invert y-axis so the highest rank (1) is at the top
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Mengambil data dari URL
url = "https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/BTC-USD-5Tahun2023.csv"
data = pd.read_csv(url)

# Mengubah kolom 'Date' menjadi tipe data datetime
data['Date'] = pd.to_datetime(data['Date'])

# Mengatur 'Date' sebagai indeks
data.set_index('Date', inplace=True)

# -------------------------------------------------------------------
# 1. Rolling Mean dan Rolling Standard Deviation
# -------------------------------------------------------------------

# Menghitung rolling mean dan rolling std untuk window 30 hari
data['Rolling_Mean'] = data['Close'].rolling(window=30).mean()
data['Rolling_Std'] = data['Close'].rolling(window=30).std()

# Plot Rolling Mean dan Rolling Standard Deviation
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Harga Close', color='blue', alpha=0.5)
plt.plot(data.index, data['Rolling_Mean'], label='Rolling Mean (30 hari)', color='orange', linestyle='--')
plt.plot(data.index, data['Rolling_Std'], label='Rolling Std (30 hari)', color='red', linestyle='--')
plt.title('Harga Bitcoin, Rolling Mean, dan Rolling Standard Deviation')
plt.xlabel('Tanggal')
plt.ylabel('Harga Close')
plt.legend(loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# -------------------------------------------------------------------
# 2. Visualisasi Perbandingan Harga Open, Close, High, Low
# -------------------------------------------------------------------

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Open'], label='Open Price', color='green', alpha=0.7)
plt.plot(data.index, data['Close'], label='Close Price', color='blue', alpha=0.7)
plt.plot(data.index, data['High'], label='High Price', color='red', alpha=0.7)
plt.plot(data.index, data['Low'], label='Low Price', color='orange', alpha=0.7)
plt.title('Perbandingan Harga Open, Close, High, Low Bitcoin')
plt.xlabel('Tanggal')
plt.ylabel('Harga')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# -------------------------------------------------------------------
# 3. Visualisasi Regresi Linear untuk Tren Harga Jangka Panjang
# -------------------------------------------------------------------

# Membuat data untuk regresi linear
X = np.array((data.index - data.index.min()).days).reshape(-1, 1)  # Menggunakan hari sejak awal sebagai fitur
y = data['Close'].values

# Melakukan regresi linear
reg = LinearRegression().fit(X, y)

# Prediksi harga berdasarkan model
predicted_price = reg.predict(X)

# Plotting regresi linear
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Harga Close', color='blue')
plt.plot(data.index, predicted_price, label='Regresi Linear', color='red', linestyle='--')
plt.title('Tren Harga Bitcoin dengan Regresi Linear')
plt.xlabel('Tanggal')
plt.ylabel('Harga Close')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# -------------------------------------------------------------------
# 4. Menghitung Value at Risk (VaR)
# -------------------------------------------------------------------

# Menghitung perubahan harga harian dalam persentase
data['Pct_Change'] = data['Close'].pct_change()

# Menghitung VaR 1% (nilai terburuk dalam 1% dari waktu)
VaR_1_percent = np.percentile(data['Pct_Change'].dropna(), 1)

# Visualisasi VaR
plt.figure(figsize=(10, 6))
plt.hist(data['Pct_Change'].dropna(), bins=50, color='purple', alpha=0.7)
plt.axvline(VaR_1_percent, color='red', linestyle='--', label=f'VaR 1%: {VaR_1_percent:.4f}')
plt.title('Distribusi Perubahan Harga Harian Bitcoin dan Value at Risk (VaR)')
plt.xlabel('Perubahan Harga Harian (%)')
plt.ylabel('Frekuensi')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# Menampilkan nilai VaR
print(f"Value at Risk (VaR) pada tingkat 1%: {VaR_1_percent:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Mengambil data dari URL
url = "https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/BTC-USD-5Tahun2023.csv"
data = pd.read_csv(url)

# Mengubah kolom 'Date' menjadi tipe data datetime
data['Date'] = pd.to_datetime(data['Date'])

# Mengatur 'Date' sebagai indeks
data.set_index('Date', inplace=True)

# -------------------------------------------------------------------
# 1. Rolling Average dan Moving Standard Deviation
# -------------------------------------------------------------------

# Menghitung rolling average dan rolling standard deviation selama 30 hari
data['Rolling_Avg'] = data['Close'].rolling(window=30).mean()
data['Rolling_Std'] = data['Close'].rolling(window=30).std()

# Visualisasi harga close dengan rolling average dan moving standard deviation
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Harga Close', color='blue', alpha=0.5)
plt.plot(data.index, data['Rolling_Avg'], label='Rolling Average (30 hari)', color='orange', linestyle='--')
plt.plot(data.index, data['Rolling_Std'], label='Rolling Std (30 hari)', color='red', linestyle='--')
plt.title('Harga Bitcoin, Rolling Average dan Rolling Standard Deviation')
plt.xlabel('Tanggal')
plt.ylabel('Harga')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# -------------------------------------------------------------------
# 2. Korelasi Antar Kolom dengan Heatmap
# -------------------------------------------------------------------

# Menghitung korelasi antar kolom
correlation_matrix = data.corr()

# Visualisasi korelasi menggunakan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Kolom')
plt.show()

# -------------------------------------------------------------------
# 3. Moving Average Convergence Divergence (MACD)
# -------------------------------------------------------------------

# Menghitung MACD
short_window = 12  # Periode untuk rata-rata cepat
long_window = 26   # Periode untuk rata-rata lambat
signal_window = 9  # Periode untuk sinyal

# Menghitung rata-rata cepat dan lambat
data['EMA12'] = data['Close'].ewm(span=short_window, adjust=False).mean()
data['EMA26'] = data['Close'].ewm(span=long_window, adjust=False).mean()

# Menghitung MACD dan sinyal MACD
data['MACD'] = data['EMA12'] - data['EMA26']
data['Signal'] = data['MACD'].ewm(span=signal_window, adjust=False).mean()

# Visualisasi MACD
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['MACD'], label='MACD', color='blue')
plt.plot(data.index, data['Signal'], label='Signal', color='orange', linestyle='--')
plt.title('Moving Average Convergence Divergence (MACD)')
plt.xlabel('Tanggal')
plt.ylabel('Nilai')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# -------------------------------------------------------------------
# 4. Visualisasi Perbandingan Kenaikan dan Penurunan Harga Bitcoin
# -------------------------------------------------------------------

# Menghitung perubahan harga harian dalam persen
data['Pct_Change'] = data['Close'].pct_change() * 100

# Menyaring kenaikan dan penurunan harga
increase = data[data['Pct_Change'] > 0]
decrease = data[data['Pct_Change'] < 0]

# Visualisasi kenaikan dan penurunan harga
plt.figure(figsize=(10, 6))
plt.plot(increase.index, increase['Pct_Change'], 'g^', label='Kenaikan Harga', alpha=0.5)
plt.plot(decrease.index, decrease['Pct_Change'], 'rv', label='Penurunan Harga', alpha=0.5)
plt.title('Perbandingan Kenaikan dan Penurunan Harga Bitcoin')
plt.xlabel('Tanggal')
plt.ylabel('Perubahan Harga (%)')
plt.legend(loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Mengambil data dari URL
url = "https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/BTC-USD-5Tahun2023.csv"
data = pd.read_csv(url)

# Mengubah kolom 'Date' menjadi tipe data datetime
data['Date'] = pd.to_datetime(data['Date'])

# Mengatur 'Date' sebagai indeks
data.set_index('Date', inplace=True)

# -------------------------------------------------------------------
# 1. Perhitungan Perubahan Harga Harian dengan NumPy
# -------------------------------------------------------------------

# Menghitung perubahan harga harian menggunakan NumPy
daily_returns = np.diff(data['Close']) / data['Close'][:-1]  # Perubahan harga harian dalam persen
daily_returns = np.concatenate(([0], daily_returns))  # Menambahkan 0 di awal untuk mencocokkan panjang data

# Menambahkan perubahan harga harian ke DataFrame
data['Daily_Returns'] = daily_returns * 100  # Mengalikan dengan 100 untuk persen

# Visualisasi perubahan harga harian
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Daily_Returns'], label='Perubahan Harga Harian (%)', color='orange')
plt.title('Perubahan Harga Harian Bitcoin')
plt.xlabel('Tanggal')
plt.ylabel('Perubahan Harga (%)')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# -------------------------------------------------------------------
# 2. Statistik Deskriptif dengan Pandas
# -------------------------------------------------------------------

# Menghitung statistik deskriptif untuk kolom harga
statistics = data[['Close', 'Daily_Returns']].describe()

print("Statistik Deskriptif:")
print(statistics)

# -------------------------------------------------------------------
# 3. Menambah Kolom Baru untuk Rata-Rata Bergerak (Moving Average) dengan NumPy
# -------------------------------------------------------------------

# Menggunakan NumPy untuk menghitung rata-rata bergerak 30 hari
moving_avg = np.convolve(data['Close'], np.ones(30)/30, mode='valid')

# Menambahkan rata-rata bergerak ke DataFrame
data['30_Day_MA'] = np.concatenate(([np.nan] * (30-1), moving_avg))  # Menambahkan NaN untuk menyesuaikan panjang data

# Visualisasi harga dan rata-rata bergerak
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Harga Bitcoin (Close)', color='blue', alpha=0.6)
plt.plot(data.index, data['30_Day_MA'], label='Rata-Rata Bergerak 30 Hari', color='green', linestyle='--')
plt.title('Harga Bitcoin dan Rata-Rata Bergerak 30 Hari')
plt.xlabel('Tanggal')
plt.ylabel('Harga (USD)')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
